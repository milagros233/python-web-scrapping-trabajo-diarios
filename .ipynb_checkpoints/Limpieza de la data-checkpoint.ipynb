```python
import pandas as pd
import Extraccion_data as ed
import numpy as np
```

```python
#puesto y cantidad de paginas a buscar
df= ed.web_scraping("administrador",5)
```

```python
df
```

```python
df.dtypes
```

```python
#la columna Empresa  y Lugar tiene al inicio y final /n y que eliminar esos caracteres
df['Empresa']=df['Empresa'].str.strip()
df['Lugar']=df['Lugar'].str.strip()
```

```python
#tolas las columnas seran en tipo oracion
df['Puesto']= df['Puesto'].str.capitalize()
df['Empresa']=df['Empresa'].str.capitalize()
df['Lugar']=df['Lugar'].str.capitalize()
df['Tipo contrato']=df['Tipo contrato'].str.capitalize()
df['Nivel de experiencia']=df['Nivel de experiencia'].str.capitalize()
df['Nivel de estudio']=df['Nivel de estudio'].str.capitalize()
df['Genero Requerido']=df['Genero Requerido'].str.capitalize()
df['Edad Requerida']=df['Edad Requerida'].str.capitalize()
```

```python
#la columna salario se convierte en float
df['Salario']=df['Salario'].apply(lambda x: float((x.split()[1]).replace(',','').replace('especifica','nan')))
```

```python
df
```

```python
df['Lugar departamento']=df['Lugar'].apply(lambda x: (x.split(",")[1]).strip().capitalize())
```

```python
df['Lugar provincia']=df['Lugar'].apply(lambda x: (x.split(",")[0]).split('/')[0].split('-')[0])
```

```python
df.drop( 'Lugar',axis=1,inplace=True)
```

```python
df.to_csv('trabajos-administracion.csv')
```
